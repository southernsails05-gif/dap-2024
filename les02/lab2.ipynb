### Задача 1: Two Sum
**Ссылка:** https://leetcode.com/problems/two-sum/

**Условие:**  
Для заданного массива целых чисел `nums` и целого числа `target`, верните индексы двух чисел, сумма которых равна `target`.

```python
class Solution1(object):
    def twoSum(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: List[int]
        """
        num_map = {}  # Храним числа и их индексы

        for i, num in enumerate(nums):
            complement = target - num

            # Если комплемент уже встречался, возвращаем индексы
            if complement in num_map:
                return [num_map[complement], i]

            # Сохраняем текущее число и его индекс
            num_map[num] = i

        return []  # Если решение не найдено

# ========== ТЕСТИРОВАНИЕ ==========
def test_twoSum():
    """Тесты для задачи Two Sum"""
    print("=" * 50)
    print("ЗАДАЧА 1: TWO SUM")
    print("=" * 50)

    solution = Solution1()
    tests = [
        {
            "nums": [2, 7, 11, 15],
            "target": 9,
            "expected": [0, 1]
        },
        {
            "nums": [3, 2, 4],
            "target": 6,
            "expected": [1, 2]
        },
        {
            "nums": [3, 3],
            "target": 6,
            "expected": [0, 1]
        }
    ]

    for i, test in enumerate(tests, 1):
        result = solution.twoSum(test["nums"], test["target"])
        status = "✓ ПРОЙДЕН" if sorted(result) == sorted(test["expected"]) else "✗ НЕ ПРОЙДЕН"
        print(f"Тест {i}: {status}")
        print(f"  Вход: nums={test['nums']}, target={test['target']}")
        print(f"  Ожидалось: {test['expected']}")
        print(f"  Получено: {result}")
        print()

test_twoSum()
```

```python
class Solution2(object):
    def isPalindrome(self, x):
        """
        :type x: int
        :rtype: bool
        """
        # Отрицательные числа не могут быть палиндромами
        if x < 0:
            return False

        # Числа, оканчивающиеся на 0 (кроме 0) не могут быть палиндромами
        if x % 10 == 0 and x != 0:
            return False

        original = x
        reversed_num = 0

        # Переворачиваем половину числа
        while x > reversed_num:
            digit = x % 10
            reversed_num = reversed_num * 10 + digit
            x //= 10

        # Для чисел с четным количеством цифр: x == reversed_num
        # Для нечетных: x == reversed_num // 10
        return x == reversed_num or x == reversed_num // 10

# ========== ТЕСТИРОВАНИЕ ==========
def test_isPalindrome():
    """Тесты для задачи Palindrome Number"""
    print("\n" + "=" * 50)
    print("ЗАДАЧА 2: PALINDROME NUMBER")
    print("=" * 50)

    solution = Solution2()
    tests = [
        {"input": 121, "expected": True},
        {"input": -121, "expected": False},
        {"input": 10, "expected": False},
        {"input": 0, "expected": True},
        {"input": 12321, "expected": True},
        {"input": 123321, "expected": True}
    ]

    for i, test in enumerate(tests, 1):
        result = solution.isPalindrome(test["input"])
        status = "✓ ПРОЙДЕН" if result == test["expected"] else "✗ НЕ ПРОЙДЕН"
        print(f"Тест {i}: {status}")
        print(f"  Вход: {test['input']}")
        print(f"  Ожидалось: {test['expected']}")
        print(f"  Получено: {result}")
        print()

test_isPalindrome()
```

### Задача 3: Roman to Integer
**Ссылка:** https://leetcode.com/problems/roman-to-integer/description/

**Условие:**  
Преобразуйте римское число в целое.

**Правила римских чисел:**  
- I = 1, V = 5, X = 10, L = 50, C = 100, D = 500, M = 1000
- Если меньшая цифра стоит перед большей, она вычитается (IV = 4, IX = 9)
- Если меньшая цифра стоит после большей, она прибавляется (VI = 6, XI = 11)


```python
class Solution3(object):
    def romanToInt(self, s):
        """
        :type s: str
        :rtype: int
        """
        roman_values = {
            'I': 1, 'V': 5, 'X': 10, 'L': 50,
            'C': 100, 'D': 500, 'M': 1000
        }

        total = 0
        prev_value = 0

        # Обрабатываем строку с конца
        for char in reversed(s):
            current_value = roman_values[char]

            # Если текущее значение меньше предыдущего, вычитаем
            if current_value < prev_value:
                total -= current_value
            else:
                total += current_value

            prev_value = current_value

        return total

# ========== ТЕСТИРОВАНИЕ ==========
def test_romanToInt():
    """Тесты для задачи Roman to Integer"""
    print("\n" + "=" * 50)
    print("ЗАДАЧА 3: ROMAN TO INTEGER")
    print("=" * 50)

    solution = Solution3()
    tests = [
        {"input": "III", "expected": 3},
        {"input": "IV", "expected": 4},
        {"input": "IX", "expected": 9},
        {"input": "LVIII", "expected": 58},
        {"input": "MCMXCIV", "expected": 1994},
        {"input": "MMXXIII", "expected": 2023}
    ]

    for i, test in enumerate(tests, 1):
        result = solution.romanToInt(test["input"])
        status = "✓ ПРОЙДЕН" if result == test["expected"] else "✗ НЕ ПРОЙДЕН"
        print(f"Тест {i}: {status}")
        print(f"  Вход: '{test['input']}'")
        print(f"  Ожидалось: {test['expected']}")
        print(f"  Получено: {result}")
        print()

test_romanToInt()
```

### Задача 4: Longest Common Prefix
**Ссылка:** https://leetcode.com/problems/longest-common-prefix/description/

**Условие:**  
Найдите самый длинный общий префикс среди массива строк.

**Пример:**  
Для ["flower","flow","flight"] результат: "fl"  
Для ["dog","racecar","car"] результат: ""


```python
class Solution4(object):
    def longestCommonPrefix(self, strs):
        """
        :type strs: List[str]
        :rtype: str
        """
        if not strs:  # Если массив пустой
            return ""

        # Берем первую строку как начальный префикс
        prefix = strs[0]

        # Сравниваем префикс с каждой следующей строкой
        for string in strs[1:]:
            # Укорачиваем префикс, пока он не станет началом строки
            while not string.startswith(prefix):
                prefix = prefix[:-1]

                # Если префикс стал пустым, общего префикса нет
                if not prefix:
                    return ""

        return prefix

# ========== ТЕСТИРОВАНИЕ ==========
def test_longestCommonPrefix():
    """Тесты для задачи Longest Common Prefix"""
    print("\n" + "=" * 50)
    print("ЗАДАЧА 4: LONGEST COMMON PREFIX")
    print("=" * 50)

    solution = Solution4()
    tests = [
        {
            "input": ["flower", "flow", "flight"],
            "expected": "fl"
        },
        {
            "input": ["dog", "racecar", "car"],
            "expected": ""
        },
        {
            "input": ["interspecies", "interstellar", "interstate"],
            "expected": "inters"
        },
        {
            "input": ["test", "test", "test"],
            "expected": "test"
        },
        {
            "input": [],
            "expected": ""
        }
    ]

    for i, test in enumerate(tests, 1):
        result = solution.longestCommonPrefix(test["input"])
        status = "✓ ПРОЙДЕН" if result == test["expected"] else "✗ НЕ ПРОЙДЕН"
        print(f"Тест {i}: {status}")
        print(f"  Вход: {test['input']}")
        print(f"  Ожидалось: '{test['expected']}'")
        print(f"  Получено: '{result}'")
        print()

test_longestCommonPrefix()
```

### Задача 5: Merge Two Sorted Lists
**Ссылка:** https://leetcode.com/problems/merge-two-sorted-lists/description/

**Условие:**  
Объедините два отсортированных связанных списка в один отсортированный список.

**Особенность:**  
Нужно работать с узлами (`ListNode`), а не преобразовывать в массивы.

**Алгоритмическая сложность:** O(n + m)  
**Сложность по памяти:** O(1)

```python
# Определение класса ListNode для задачи
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution5(object):
    def mergeTwoLists(self, list1, list2):
        """
        :type list1: Optional[ListNode]
        :type list2: Optional[ListNode]
        :rtype: Optional[ListNode]
        """
        # Создаем фиктивную голову для удобства
        dummy = ListNode(0)
        current = dummy

        # Пока оба списка не пусты
        while list1 and list2:
            if list1.val <= list2.val:
                current.next = list1
                list1 = list1.next
            else:
                current.next = list2
                list2 = list2.next
            current = current.next

        # Присоединяем оставшуюся часть непустого списка
        if list1:
            current.next = list1
        else:
            current.next = list2

        return dummy.next

# ========== ВСПОМОГАТЕЛЬНЫЕ ФУНКЦИИ ==========
def create_list(arr):
    """Создает связанный список из массива"""
    if not arr:
        return None

    head = ListNode(arr[0])
    current = head

    for val in arr[1:]:
        current.next = ListNode(val)
        current = current.next

    return head

def list_to_array(head):
    """Преобразует связанный список в массив"""
    arr = []
    while head:
        arr.append(head.val)
        head = head.next
    return arr

# ========== ТЕСТИРОВАНИЕ ==========
def test_mergeTwoLists():
    """Тесты для задачи Merge Two Sorted Lists"""
    print("\n" + "=" * 50)
    print("ЗАДАЧА 5: MERGE TWO SORTED LISTS")
    print("=" * 50)

    solution = Solution5()
    tests = [
        {
            "list1": [1, 2, 4],
            "list2": [1, 3, 4],
            "expected": [1, 1, 2, 3, 4, 4]
        },
        {
            "list1": [],
            "list2": [0],
            "expected": [0]
        },
        {
            "list1": [1, 3, 5],
            "list2": [2, 4, 6],
            "expected": [1, 2, 3, 4, 5, 6]
        }
    ]

    for i, test in enumerate(tests, 1):
        list1 = create_list(test["list1"])
        list2 = create_list(test["list2"])

        result = solution.mergeTwoLists(list1, list2)
        result_array = list_to_array(result)

        status = "✓ ПРОЙДЕН" if result_array == test["expected"] else "✗ НЕ ПРОЙДЕН"
        print(f"Тест {i}: {status}")
        print(f"  Вход: list1={test['list1']}, list2={test['list2']}")
        print(f"  Ожидалось: {test['expected']}")
        print(f"  Получено: {result_array}")
        print()

test_mergeTwoLists()
```

### Задача 6: Longest Substring Without Repeating Characters
**Ссылка:** https://leetcode.com/problems/longest-substring-without-repeating-characters/description/

**Условие:**  
Найдите длину самой длинной подстроки без повторяющихся символов.

**Пример:**  
"abcabcbb" → 3 ("abc")  
"bbbbb" → 1 ("b")

**Алгоритмическая сложность:** O(n)  
**Сложность по памяти:** O(min(n, m)), где m - размер алфавита

```python
class Solution6(object):
    def lengthOfLongestSubstring(self, s):
        """
        :type s: str
        :rtype: int
        """
        char_index = {}  # Словарь для хранения последнего индекса символа
        max_length = 0
        start = 0  # Начало текущей подстроки

        for end, char in enumerate(s):
            # Если символ уже встречался и его индекс >= start
            if char in char_index and char_index[char] >= start:
                # Сдвигаем начало окна
                start = char_index[char] + 1

            # Обновляем индекс символа
            char_index[char] = end

            # Обновляем максимальную длину
            current_length = end - start + 1
            if current_length > max_length:
                max_length = current_length

        return max_length

# ========== ТЕСТИРОВАНИЕ ==========
def test_lengthOfLongestSubstring():
    """Тесты для задачи Longest Substring"""
    print("\n" + "=" * 50)
    print("ЗАДАЧА 6: LONGEST SUBSTRING WITHOUT REPEATING CHARACTERS")
    print("=" * 50)

    solution = Solution6()
    tests = [
        {"input": "abcabcbb", "expected": 3},
        {"input": "bbbbb", "expected": 1},
        {"input": "pwwkew", "expected": 3},
        {"input": "", "expected": 0},
        {"input": "abcdef", "expected": 6},
        {"input": "aab", "expected": 2}
    ]

    for i, test in enumerate(tests, 1):
        result = solution.lengthOfLongestSubstring(test["input"])
        status = "✓ ПРОЙДЕН" if result == test["expected"] else "✗ НЕ ПРОЙДЕН"
        print(f"Тест {i}: {status}")
        print(f"  Вход: '{test['input']}'")
        print(f"  Ожидалось: {test['expected']}")
        print(f"  Получено: {result}")
        print()

test_lengthOfLongestSubstring()
```
