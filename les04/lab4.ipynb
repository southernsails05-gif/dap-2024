```python
# Лабораторная работа №4
## Анализ данных Titanic с использованием Pandas

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Настройка отображения
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)

print("Библиотеки загружены успешно")
print(f"Версия Pandas: {pd.__version__}")
print("=" * 70)
```

```python
# ========== ЗАДАЧА 1 ==========
print("ЗАДАЧА 1: Создание DataFrame из файла titanic.csv")
print("-" * 50)

# Способ 1: Загрузка из интернета (рекомендуется)
try:
    # URL датасета Titanic
    url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
    df = pd.read_csv(url, index_col='PassengerId')
    print("✓ Данные загружены из интернета")

except:
    # Способ 2: Загрузка локального файла через Colab
    print("Загружаем файл через Colab...")

    # Импортируем модуль для загрузки файлов
    from google.colab import files

    # Загружаем файл
    uploaded = files.upload()

    # Ищем загруженный файл
    for filename in uploaded.keys():
        if 'titanic' in filename.lower():
            df = pd.read_csv(filename, index_col='PassengerId')
            print(f"✓ Файл '{filename}' загружен успешно")
            break
    else:
        # Если файл не найден, используем встроенный датасет
        print("Используем встроенный датасет Seaborn...")
        import seaborn as sns
        df = sns.load_dataset('titanic')
        df = df.set_index(df.index)  # Создаем числовой индекс
        df.index.name = 'PassengerId'
        print("✓ Встроенный датасет Titanic загружен")

print(f"\nДанные успешно загружены. Размер: {df.shape[0]} строк, {df.shape[1]} столбцов")
print(f"Индекс: {df.index.name}")
print("\nСтолбцы:", list(df.columns))
print("=" * 70)
```

```python
# ========== ЗАДАЧА 2 ==========
print("\nЗАДАЧА 2: Первые 6 строк данных")
print("-" * 50)

print("Первые 6 строк:")
print(df.head(6))
print("=" * 70)
```

```python
# ========== ЗАДАЧА 3 ==========
print("\nЗАДАЧА 3: Описание данных")
print("-" * 50)

print("Информация о DataFrame:")
print(df.info())

print("\n\nСтатистическое описание числовых столбцов:")
print(df.describe())

# Проверяем, есть ли нечисловые столбвы
non_numeric_cols = df.select_dtypes(include=['object']).columns
if len(non_numeric_cols) > 0:
    print(f"\n\nСтатистическое описание нечисловых столбцов ({len(non_numeric_cols)}):")
    print(df[non_numeric_cols].describe())
else:
    print("\nНет нечисловых столбцов для описания")

print("=" * 70)
```

```python
# ========== ЗАДАЧА 4 ==========
print("\nЗАДАЧА 4: Сколько мужчин/женщин находилось на борту?")
print("-" * 50)

gender_counts = df['Sex'].value_counts()
print("Распределение по полу:")
print(gender_counts)

print(f"\n✓ Всего мужчин: {gender_counts.get('male', 0)}")
print(f"✓ Всего женщин: {gender_counts.get('female', 0)}")
print("=" * 70)
```

```python
# ========== ЗАДАЧА 5 ==========
print("\nЗАДАЧА 5: Распределение переменной Pclass")
print("-" * 50)

# Общее распределение
print("1. ОБЩЕЕ распределение по классам:")
pclass_all = df['Pclass'].value_counts().sort_index()
print(pclass_all)

# По полу отдельно
print("\n2. Для МУЖЧИН:")
men_class = df[df['Sex'] == 'male']['Pclass'].value_counts().sort_index()
print(men_class)

print("\n3. Для ЖЕНЩИН:")
women_class = df[df['Sex'] == 'female']['Pclass'].value_counts().sort_index()
print(women_class)

# Ответ на конкретный вопрос
men_2nd = men_class.get(2, 0)
print(f"\n✓ Мужчин 2-го класса было: {men_2nd}")
print("=" * 70)
```

```python
# ========== ЗАДАЧА 6 ==========
print("\nЗАДАЧА 6: Медиана и стандартное отклонение платежей (Fare)")
print("-" * 50)

fare_median = round(df['Fare'].median(), 2)
fare_std = round(df['Fare'].std(), 2)

print(f"✓ Медиана платежа: ${fare_median}")
print(f"✓ Среднеквадратичное отклонение: ${fare_std}")

# Дополнительная информация
print(f"\nДополнительно:")
print(f"• Минимальный платёж: ${df['Fare'].min():.2f}")
print(f"• Максимальный платёж: ${df['Fare'].max():.2f}")
print(f"• Средний платёж: ${df['Fare'].mean():.2f}")
print("=" * 70)
```

```python
# ========== ЗАДАЧА 7 ==========
print("\nЗАДАЧА 7: Выживаемость людей <30 лет vs >60 лет")
print("-" * 50)

# Создаем группы (игнорируем NaN в возрасте)
young = df[df['Age'] < 30]
old = df[df['Age'] > 60]

# Доли выживших
young_rate = young['Survived'].mean()
old_rate = old['Survived'].mean()

print(f"Пассажиров младше 30 лет: {len(young)} чел.")
print(f"Пассажиров старше 60 лет: {len(old)} чел.")
print(f"\nДоля выживших <30 лет: {young_rate:.2%} ({young_rate:.3f})")
print(f"Доля выживших >60 лет: {old_rate:.2%} ({old_rate:.3f})")

# Ответ
if young_rate > old_rate:
    print(f"\n✓ ПРАВДА: Молодые выживали чаще ({young_rate:.1%} > {old_rate:.1%})")
else:
    print(f"\n✗ ЛОЖЬ: Молодые выживали реже ({young_rate:.1%} < {old_rate:.1%})")
print("=" * 70)
```

```python
# ========== ЗАДАЧА 8 ==========
print("\nЗАДАЧА 8: Выживаемость женщин vs мужчин")
print("-" * 50)

# Группируем по полу и вычисляем долю выживших
survival_by_sex = df.groupby('Sex')['Survived'].mean()

print("Доля выживших по полу:")
for sex, rate in survival_by_sex.items():
    print(f"• {sex}: {rate:.2%} ({rate:.3f})")

# Конкретные числа
women_rate = survival_by_sex.get('female', 0)
men_rate = survival_by_sex.get('male', 0)

if women_rate > men_rate:
    print(f"\n✓ ПРАВДА: Женщины выживали чаще ({women_rate:.1%} > {men_rate:.1%})")
else:
    print(f"\n✗ ЛОЖЬ: Женщины выживали реже")
print("=" * 70)
```

```python
# ========== ЗАДАЧА 9 ==========
print("\nЗАДАЧА 9: Самое популярное имя среди мужчин")
print("-" * 50)

# Извлекаем имена из столбца Name
# Имя обычно идет после запятой и перед точкой: "Braund, Mr. Owen Harris"
def extract_first_name(name):
    try:
        # Пытаемся извлечь часть между запятой и точкой
        parts = name.split(', ')
        if len(parts) > 1:
            title_part = parts[1]
            # Убираем титул (Mr., Mrs., Miss и т.д.)
            name_parts = title_part.split('. ')
            if len(name_parts) > 1:
                return name_parts[1].split()[0]  # Первое имя
    except:
        pass
    return None

# Применяем функцию ко всем именам
df['FirstName'] = df['Name'].apply(extract_first_name)

# Считаем популярные имена среди мужчин
male_names = df[df['Sex'] == 'male']['FirstName'].dropna()
top_male_name = male_names.mode()

if len(top_male_name) > 0:
    print(f"✓ Самое популярное мужское имя: '{top_male_name.iloc[0]}'")
    print(f"  Встречается {male_names.value_counts().iloc[0]} раз")
else:
    print("Не удалось определить самое популярное имя")

# Покажем топ-5 имен
print("\nТоп-5 мужских имен:")
print(male_names.value_counts().head())
print("=" * 70)
```

```python
# ========== ЗАДАЧА 10 ==========
print("\nЗАДАЧА 10: Средний возраст по полу и классу")
print("-" * 50)

# Средний возраст для каждого сочетания пол-класс
age_by_sex_class = df.groupby(['Sex', 'Pclass'])['Age'].mean().unstack()

print("Средний возраст (лет):")
print(age_by_sex_class.round(1))

# Проверка утверждений
print("\nПРОВЕРКА УТВЕРЖДЕНИЙ:")

# 1. Мужчины 1-го класса старше 40?
men_1st_age = age_by_sex_class.loc['male', 1]
print(f"1. Мужчины 1-го класса в среднем старше 40 лет?")
print(f"   Средний возраст: {men_1st_age:.1f} лет → {'ДА' if men_1st_age > 40 else 'НЕТ'}")

# 2. Женщины 1-го класса старше 40?
women_1st_age = age_by_sex_class.loc['female', 1]
print(f"\n2. Женщины 1-го класса в среднем старше 40 лет?")
print(f"   Средний возраст: {women_1st_age:.1f} лет → {'ДА' if women_1st_age > 40 else 'НЕТ'}")

# 3. Мужчины старше женщин в том же классе?
print(f"\n3. Мужчины всех классов в среднем старше женщин того же класса?")
for pclass in [1, 2, 3]:
    men_age = age_by_sex_class.loc['male', pclass]
    women_age = age_by_sex_class.loc['female', pclass]
    print(f"   Класс {pclass}: муж {men_age:.1f} vs жен {women_age:.1f} → "
          f"{'ДА' if men_age > women_age else 'НЕТ'}")

# 4. Возраст уменьшается с классом?
print(f"\n4. Люди в 1 классе старше, чем во 2-ом, а те старше 3-го класса?")
avg_age_by_class = df.groupby('Pclass')['Age'].mean()
print(f"   Средний возраст по классам:")
for pclass, age in avg_age_by_class.items():
    print(f"   Класс {pclass}: {age:.1f} лет")

# Проверка убывания
is_decreasing = (avg_age_by_class[1] > avg_age_by_class[2] > avg_age_by_class[3])
print(f"   Утверждение верно? {'ДА' if is_decreasing else 'НЕТ'}")

print("\n5. Средний возраст у спасенных и погибших:")
survival_age = df.groupby('Survived')['Age'].mean()
print(f"   • Выжившие: {survival_age.get(1, 0):.1f} лет")
print(f"   • Погибшие: {survival_age.get(0, 0):.1f} лет")

if survival_age.get(0, 0) > survival_age.get(1, 0):
    print(f"   ✓ Средний возраст погибших ВЫШЕ")
else:
    print(f"   ✗ Средний возраст погибших НИЖЕ")
print("=" * 70)
```
