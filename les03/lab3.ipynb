Освоение основных операций работы с массивами NumPy: создание, индексация, модификация, математические операции.

```python
import numpy as np

# Проверяем версию NumPy
print(f"Версия NumPy: {np.__version__}")
print("=" * 60)
```

```python
# ========== ЗАДАЧА 1 ==========
print("ЗАДАЧА 1: Создать массив 10x10 со случайными значениями, найти минимум и максимум")
print("-" * 50)

# Создаем массив 10x10 со случайными значениями от 0 до 100
array_10x10 = np.random.rand(10, 10) * 100
print(f"Массив 10x10:\n{array_10x10}")
print(f"\nМинимальное значение: {array_10x10.min():.4f}")
print(f"Максимальное значение: {array_10x10.max():.4f}")
print(f"Среднее значение: {array_10x10.mean():.4f}")
print("=" * 60)
```

```python
# ========== ЗАДАЧА 2 ==========
print("\nЗАДАЧА 2: Создать 8x8 матрицу в шахматном порядке")
print("-" * 50)

# Способ 1: Используя срезы
chess_board = np.zeros((8, 8), dtype=int)
chess_board[1::2, ::2] = 1  # Нечетные строки, четные столбцы
chess_board[::2, 1::2] = 1  # Четные строки, нечетные столбцы

print("Шахматная доска 8x8:")
print(chess_board)
print("=" * 60)
```

```python
# ========== ЗАДАЧА 3 ==========
print("\nЗАДАЧА 3: Поменять знак у элементов, значения которых между 3 и 8")
print("-" * 50)

z = np.arange(11)
print(f"Исходный массив: {z}")

# Меняем знак у элементов между 3 и 8 (не включая границы)
mask = (z > 3) & (z < 8)
z[mask] = -z[mask]

print(f"Результат:       {z}")
print("=" * 60)
```

```python
# ========== ЗАДАЧА 4 ==========
print("\nЗАДАЧА 4: Найти среднее значение строки и вычесть его из каждого элемента")
print("-" * 50)

z = np.random.rand(5, 10)
print(f"Исходная матрица (5x10):\n{z}")

# Находим среднее по строкам (axis=1)
row_means = z.mean(axis=1, keepdims=True)  # keepdims сохраняет размерность
print(f"\nСредние значения строк: {row_means.flatten()}")

# Вычитаем среднее из каждой строки
z_normalized = z - row_means
print(f"\nМатрица после вычитания среднего:\n{z_normalized}")

# Проверка: среднее каждой строки должно быть ~0
print(f"\nПроверка (средние после нормализации): {z_normalized.mean(axis=1).round(10)}")
print("=" * 60)
```

```python
# ========== ЗАДАЧА 5 ==========
print("\nЗАДАЧА 5: Поменять местами 1 и 2 строки матрицы")
print("-" * 50)

z = np.arange(25).reshape(5,5)
print(f"Исходная матрица 5x5:\n{z}")

# Меняем строки 1 и 2 (индексы 0 и 1)
z[[0, 1]] = z[[1, 0]]
print(f"\nМатрица после замены строк 1 и 2:\n{z}")
print("=" * 60)
```

```python
# ========== ЗАДАЧА 6 ==========
print("\nЗАДАЧА 6: Преобразовать декартовы координаты в полярные")
print("-" * 50)

z = np.random.random((10,2)) * 10  # 10 точек, координаты от 0 до 10
print(f"Декартовы координаты (10 точек):\n{z}")

# Преобразование в полярные координаты
# r = sqrt(x² + y²), θ = arctan2(y, x)
x = z[:, 0]
y = z[:, 1]

r = np.sqrt(x**2 + y**2)  # Радиус
theta = np.arctan2(y, x)  # Угол в радианах

# Собираем результат
polar_coords = np.column_stack((r, theta))
print(f"\nПолярные координаты (радиус, угол в радианах):\n{polar_coords}")
print("=" * 60)
```

```python
# ========== ЗАДАЧА 7 ==========
print("\nЗАДАЧА 7: Заменить максимальный элемент массива на ноль")
print("-" * 50)

z = np.random.random(10)
print(f"Исходный массив: {z}")

# Находим индекс максимального элемента
max_index = np.argmax(z)
print(f"Максимальный элемент: z[{max_index}] = {z[max_index]:.4f}")

# Заменяем максимальный элемент на 0
z[max_index] = 0
print(f"Результат:         {z}")
print("=" * 60)
```

```python
# ========== ЗАДАЧА 8 ==========
print("\nЗАДАЧА 8: Добавить 1 к элементам с заданными индексами (только 1 раз)")
print("-" * 50)

z = np.ones(10)
i = np.random.randint(0, len(z), 20)  # 20 случайных индексов
print(f"Исходный массив: {z}")
print(f"Индексы для добавления (могут повторяться): {i}")

# Находим уникальные индексы
unique_indices = np.unique(i)
print(f"Уникальные индексы: {unique_indices}")

# Добавляем 1 только к уникальным индексам
z[unique_indices] += 1
print(f"Результат:         {z}")
print("=" * 60)
```

```python
# ========== ЗАДАЧА 9 ==========
print("\nЗАДАЧА 9: Вставить три нуля между каждым элементом вектора")
print("-" * 50)

z = np.array([1, 2, 3, 4, 5])
print(f"Исходный вектор: {z}")

# Создаем новый массив с нулями
# Длина нового массива: 5 элементов + (5-1) * 3 нулей = 17
result = np.zeros(len(z) + (len(z) - 1) * 3, dtype=z.dtype)

# Вставляем исходные элементы на каждую 4-ю позицию
result[::4] = z

print(f"Результат:        {result}")
print("=" * 60)
```

```python
# ========== ЗАДАЧА 10 ==========
print("\nЗАДАЧА 10: Найти ближайшее к заданному значению число в массиве")
print("-" * 50)

z = np.arange(100)  # Упорядоченный массив от 0 до 99
v = np.random.uniform(0, 100)  # Случайное число от 0 до 100
print(f"Массив: от 0 до 99")
print(f"Искомое значение: v = {v:.4f}")

# Находим индекс ближайшего элемента
# Способ 1: используя argmin с абсолютной разницей
idx = np.argmin(np.abs(z - v))
nearest_value = z[idx]

print(f"Ближайшее значение: z[{idx}] = {nearest_value}")
print(f"Разница: |{v:.4f} - {nearest_value}| = {abs(v - nearest_value):.4f}")
print("=" * 60)
```
